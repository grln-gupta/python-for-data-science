```python
## Traverse a list
'''You are given a list that contains integers. You need to print the elements of the list with a space between them.
Note: Do not add a new line at the end.

Input: arr[] = [54, 43, 2, 1, 5]
Output: 54 43 2 1 5
Explanation: Just traverse and print the numbers.
'''

arr1 = [54, 43, 2, 1, 5]
for arr in arr1 :
  print(arr, sep=' ', end=' ')

```

```python
# ## Sum of a list
total_sum = 0
for index, value in enumerate(arr1):
  total_sum += value
print(f"The sum of the list is: {total_sum}")
```

```python
#Fabnocii series is the sum of the preceeding 2 number gives next number
import math

def is_perfect_sq(x):
    s = int(math.sqrt(x))
    return s * s == x

def is_fibonacci(n):
    return is_perfect_sq(5 * n * n + 4) or is_perfect_sq(5 * n * n - 4)

for i in range(1, 10):
    if is_fibonacci(i):
        print(f"{i} is a Fibonacci Number")
    else:
        print(f"{i} is not a Fibonacci Number")
```
