A list is a versatile &  mutable source it can hold a collection of data in between two [].
 ->

*   It can store any data type.
*   mutable is nothig but  values can be changed during the runtime
*   List item
*  example: my_list1 = [1,2,4, 'a']


    

```python
my_list1 = [1,2,4, 'a']
print(my_list1)
```

```python
my_list1[0] = 'read'
print(my_list1[0])
```

```python
sublist = my_list1[2:3]
print(sublist)
```

```python
print(my_list1)
my_list1.reverse()
print(my_list1)
my_list1[::-1] #-1 indicates that the values are printed in reverse order
reveredList = my_list1
print(reveredList)
```

```python

# what value that needs to removed or needs to add is given in the brackets
my_list1.append(10)
print(my_list1)
my_list1.remove('read')
print(my_list1)
```

```python
# List concentenation
my_list2 = [8,8,9,10]
myList3 = my_list1 + my_list2
print(myList3)
```

```python
#Repetetion tasks & length of list
my_list4 = [10] * 10
print(my_list4)
lengthOfList = len(my_list4)
print('')
print('length of list ', lengthOfList)
```

```python
#Nested Lists
myList6 = [1, 3, [4,5], 6,7, [8,9]] ##[1 index is 0, 3 is index 1, [4,5] is at index 2 -- inside that sublist[4,5] is at index 0 & 1]
print(myList6[2][0])
```

```python
myList6.append(33)
print('')
print('append', myList6)
print('')
myList6.insert(1, 44)
print('')
print('insert', myList6)
myList6.pop(4)
print('')
print('pop', myList6)
myList6.remove(33)
print('')
print('remove', myList6)
```

```python
## list Comprehension
squres = [y**2 for y in range(101)]
print(squres)
print('')
print('length of squares', len(squres))
```

```python
## Traverse a list
'''You are given a list that contains integers. You need to print the elements of the list with a space between them.
Note: Do not add a new line at the end.

Input: arr[] = [54, 43, 2, 1, 5]
Output: 54 43 2 1 5
Explanation: Just traverse and print the numbers.
'''

arr1 = [54, 43, 2, 1, 5]
for arr in arr1 :
  print(arr, sep=' ', end=' ')

```

```python
# ## Sum of a list
total_sum = 0
for index, value in enumerate(arr1):
  total_sum += value
print(f"The sum of the list is: {total_sum}")
```
