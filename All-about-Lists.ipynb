A list is a versatile &  mutable source it can hold a collection of data in between two [].
 ->

*   It can store any data type.
*   mutable is nothig but  values can be changed during the runtime
*   List item
*  example: my_list1 = [1,2,4, 'a']


    

```python
my_list1 = [1,2,4, 'a']
print(my_list1)
```

```python
my_list1[0] = 'read'
print(my_list1[0])
```

```python
sublist = my_list1[2:3]
print(sublist)
```

```python
print(my_list1)
my_list1.reverse()
print(my_list1)
my_list1[::-1] #-1 indicates that the values are printed in reverse order
reveredList = my_list1
print(reveredList)
```

```python

# what value that needs to removed or needs to add is given in the brackets
my_list1.append(10)
print(my_list1)
my_list1.remove('read')
print(my_list1)
```
